```python
tinfo = pd.read_csv(data/ 'row_and_column_count.csv')
tinfo = tinfo[tinfo.num_rows > 0]
tinfo
```

```python
f_to_col = {}
for idx, e in tinfo.iterrows():
    f_to_col[e.file_name] = (~pd.read_csv(data/f'{e.file_name}.csv', low_memory=False).isna()).sum() if e.num_rows > 0 else None
f_to_col
```

```python
df = pd.DataFrame(f_to_col).transpose().reset_index()
df.columns = ['file_name'] + list(df.columns[1:])
df
```

```python
(~df.isna()).sum().sort_values()[-500:]
```

```python
tinfo = df.merge(tinfo, on='file_name')
tinfo.shape
```

```python
col_count = df.drop('file_name',axis=1).sum().sort_values()
col_count
```

```python
cats = pd.cut(col_count, [0, 1e1, 1e2,1e3,1e4, 1e5, 1e6, 1e7,],
      right=True, include_lowest=True,
       labels=['0-10', '10-100', '100-1000', '1000-10000', '10000-100000',
                         '100000-1000000', '1000000-10000000'])
cats
```

```python
cats.value_counts().sort_index()
```
