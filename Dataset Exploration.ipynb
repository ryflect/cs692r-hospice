```python
# imports
import os
import numpy 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read datafiles into dataframes
# TCGA-KIRC
df_kirc = {}
nf_kirc = 0
for filename in os.listdir("./TCGAdata/TCGA-KIRC/KIRC/"):
    if filename.endswith(".txt"):
        df_kirc[nf_kirc] = pd.read_csv("./TCGAdata/TCGA-KIRC/KIRC/" + filename, sep=" ", delimiter="\t")
        nf_kirc = nf_kirc + 1
print("Total number of KIRC files: ", nf_kirc)

# TCGA-BRCA
df_brca = {}
nf_brca = 0
for filename in os.listdir("./TCGAdata/TCGA-BRCA/BRCA/"):
    if filename.endswith(".txt"):
        df_brca[nf_brca] = pd.read_csv("./TCGAdata/TCGA-BRCA/BRCA/" + filename, sep=" ", delimiter="\t")
        nf_brca = nf_brca + 1
print("Total number of BRCA files: ", nf_brca)

# TCGA-COAD
df_coad = {}
nf_coad = 0
for filename in os.listdir("./TCGAdata/TCGA-COAD/COAD/"):
    if filename.endswith(".txt"):
        df_coad[nf_coad] = pd.read_csv("./TCGAdata/TCGA-COAD/COAD/" + filename, sep=" ", delimiter="\t")
        nf_coad = nf_coad + 1
print("Total number of COAD files: ", nf_coad)

# TCGA-GBM
df_gbm = {}
nf_gbm = 0
for filename in os.listdir("./TCGAdata/TCGA-GBM/GBM/"):
    if filename.endswith(".txt"):
        df_gbm[nf_gbm] = pd.read_csv("./TCGAdata/TCGA-GBM/GBM/" + filename, sep=" ", delimiter="\t")
        nf_gbm = nf_gbm + 1
print("Total number of GBM files: ", nf_gbm)

# TCGA-KIRP
df_kirp = {}
nf_kirp = 0
for filename in os.listdir("./TCGAdata/TCGA-KIRP/KIRP/"):
    if filename.endswith(".txt"):
        df_kirp[nf_kirp] = pd.read_csv("./TCGAdata/TCGA-KIRP/KIRP/" + filename, sep=" ", delimiter="\t")
        nf_kirp = nf_kirp + 1
print("Total number of KIRP files: ", nf_kirp)

# TCGA-LUAD
df_luad = {}
nf_luad = 0
for filename in os.listdir("./TCGAdata/TCGA-LUAD/LUAD/"):
    if filename.endswith(".txt"):
        df_luad[nf_luad] = pd.read_csv("./TCGAdata/TCGA-LUAD/LUAD/" + filename, sep=" ", delimiter="\t")
        nf_luad = nf_luad + 1
print("Total number of LUAD files: ", nf_luad)

# TCGA-LUSC
df_lusc = {}
nf_lusc = 0
for filename in os.listdir("./TCGAdata/TCGA-LUSC/LUSC/"):
    if filename.endswith(".txt"):
        df_lusc[nf_lusc] = pd.read_csv("./TCGAdata/TCGA-LUSC/LUSC/" + filename, sep=" ", delimiter="\t")
        nf_lusc = nf_lusc + 1
print("Total number of LUSC files: ", nf_lusc)

# TCGA-OV
df_ov = {}
nf_ov = 0
for filename in os.listdir("./TCGAdata/TCGA-OV/OV/"):
    if filename.endswith(".txt"):
        df_ov[nf_ov] = pd.read_csv("./TCGAdata/TCGA-OV/OV/" + filename, sep=" ", delimiter="\t")
        nf_ov = nf_ov + 1
print("Total number of OV files: ", nf_ov)

# TCGA-READ
df_read = {}
nf_read = 0
for filename in os.listdir("./TCGAdata/TCGA-READ/READ/"):
    if filename.endswith(".txt"):
        df_read[nf_read] = pd.read_csv("./TCGAdata/TCGA-READ/READ/" + filename, sep=" ", delimiter="\t")
        nf_read = nf_read + 1
print("Total number of READ files: ", nf_read)

# TCGA-UCEC
df_ucec = {}
nf_ucec = 0
for filename in os.listdir("./TCGAdata/TCGA-UCEC/UCEC/"):
    if filename.endswith(".txt"):
        df_ucec[nf_ucec] = pd.read_csv("./TCGAdata/TCGA-UCEC/UCEC/" + filename, sep=" ", delimiter="\t")
        nf_ucec = nf_ucec + 1
print("Total number of UCEC files: ", nf_ucec)
```
